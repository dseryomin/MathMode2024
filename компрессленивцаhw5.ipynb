```python
def reducer(X, r):
    U, S, V = np.linalg.svd(X, full_matrices=False) # разложение SVD
    Y = U[:,:r] @ np.diag(S[:r]) @ V[:r]
    Y[Y < 0] = 0
    Y[Y > 255] = 255

    return Y
```

Функция производящая сжатие
